## Part 1 — Python Fundamentals

The six constructs you will use in *every* AI script.

```python
# 1. Variables and data types
model_name = "GPT-4"          # str (string) — text in quotes
accuracy   = 0.923            # float — a decimal number
num_params = 175_000_000_000  # int — a whole number (underscores are just for readability, like commas)
is_trained = True             # bool — True or False

print(f"Model : {model_name}")
# f-strings let you embed variables directly in text using {curly braces}
# :.1% formats a decimal as a percentage with 1 decimal place (0.923 → 92.3%)
print(f"Accuracy : {accuracy:.1%}")
# :, adds thousands separators (175000000000 → 175,000,000,000)
print(f"Parameters: {num_params:,}")
```

```python
# 2. Lists — ordered, mutable sequences (you can change them after creation)
scores = [0.82, 0.91, 0.78, 0.95, 0.88]

print("All scores :", scores)
print("First score:", scores[0])    # index 0 = first item (Python counts from 0)
print("Last score :", scores[-1])   # -1 = last item, -2 = second to last, etc.
# sorted() returns a new sorted list (the original is unchanged)
# reverse=True sorts from highest to lowest
# [:3] slices the first 3 items
print("Top 3      :", sorted(scores, reverse=True)[:3])
```

```python
# 3. Dictionaries — key-value stores (like a lookup table or a JSON object)
# Each entry is a key: value pair — keys must be unique
model_info = {
    "name"       : "ResNet-50",
    "task"       : "image classification",
    "top1_acc"   : 0.761,
    "params_M"   : 25.6
}

# .items() returns each key-value pair so we can loop through both at once
# {key:<12} means left-align the key text within a field of 12 characters (for neat columns)
for key, value in model_info.items():
    print(f"  {key:<12} -> {value}")
```

```python
# 4. Functions — reusable blocks of logic
def accuracy_grade(acc):
    """Return a letter grade for a model accuracy score."""
    if acc >= 0.95:
        return "A+"
    elif acc >= 0.90:
        return "A"
    elif acc >= 0.80:
        return "B"
    else:
        return "C"

for score in scores:
    print(f"  {score:.0%}  =>  {accuracy_grade(score)}")
```

```python
# 5. List comprehensions — a compact way to build a new list from an existing one
# Format: [expression for item in iterable if condition]

# Convert fraction accuracies to percentages
# This is equivalent to a for loop that appends round(s*100, 1) for each s in scores
pct_scores = [round(s * 100, 1) for s in scores]
print("Percentage scores:", pct_scores)

# Filter only passing scores (>= 85%) — the `if s >= 85` part acts as a filter
passing = [s for s in pct_scores if s >= 85]
print("Passing (>=85%)  :", passing)
```

```python
# 6. Classes — blueprints for creating objects with their own data and behaviour
# In AI, you'll see classes used to define neural network architectures
class MLModel:
    def __init__(self, name, task):
        # __init__ is the constructor — it runs automatically when you create a new object
        # `self` refers to the specific instance being created (like "this" in other languages)
        # self.name stores the value as an attribute on the object
        self.name = name
        self.task = task
        self.trained = False   # default starting state

    def train(self):
        self.trained = True   # change the object's state
        print(f"{self.name} has been trained for '{self.task}'.")

    def predict(self, sample):
        if not self.trained:
            return "Model not yet trained!"
        return f"Prediction on '{sample}' by {self.name}"

# Create an instance of the class (calls __init__ automatically)
my_model = MLModel("TinyBERT", "sentiment analysis")
my_model.train()
print(my_model.predict("I love this AI course!"))
```

## Part 2 — NumPy: Fast Arrays

```python
import numpy as np

# Create arrays
pixels = np.array([120, 200, 85, 255, 30, 178])  # like a row of pixel values
print("Array         :", pixels)
print("Shape         :", pixels.shape)
print("Mean          :", pixels.mean())
print("Std deviation :", pixels.std().round(2))
```

```python
# 2-D arrays — like an image or a matrix of weights
weights = np.array([
    [0.2, -0.5,  0.8],
    [1.1,  0.3, -0.9],
    [0.0,  0.7,  0.4]
])

print("Weights matrix:")
print(weights)
print("\nShape:", weights.shape)   # (rows, columns)
```

```python
# Vectorised math — NumPy operates on every element at once (much faster than a Python loop)
inputs = np.array([0.5, 1.2, -0.3])

# Matrix multiplication: @ is the operator for matrix multiply
# This is exactly what a neuron does — multiply inputs by weights and sum them up
# weights has shape (3, 3) and inputs has shape (3,) → output has shape (3,)
output = weights @ inputs
print("Neuron outputs:", output.round(3))

# Normalise pixels to 0-1 range (divides every element in the array by 255)
# This is a common preprocessing step — neural networks prefer small input values
normalised = pixels / 255.0
print("Normalised pixels:", normalised.round(3))
```

```python
# Slicing — extract sub-arrays using [row_range, column_range]
# np.arange(1, 26) creates integers from 1 to 25
# .reshape(5, 5) arranges them into a 5-row x 5-column grid
matrix = np.arange(1, 26).reshape(5, 5)
print("Full 5x5 matrix:")
print(matrix)

# matrix[:3, :3] means: take the first 3 rows AND first 3 columns
# : alone means "all", :3 means "up to (but not including) index 3"
print("\nTop-left 3x3 patch (like a CNN filter sees):")
print(matrix[:3, :3])
```

## Part 3 — pandas: Data Wrangling

```python
import pandas as pd

# pd.read_csv() loads a CSV file (from a URL or local path) into a DataFrame
# A DataFrame is like a spreadsheet — rows are samples, columns are features
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# .shape returns a tuple: (number of rows, number of columns)
print(f"Shape: {df.shape}  (rows x columns)")
# .head() shows the first 5 rows by default — useful for a quick preview
df.head()
```

```python
# Explore the dataset
print("\n--- Column types ---")
print(df.dtypes)

print("\n--- Summary statistics ---")
df.describe().round(2)
```

```python
# Count species — understanding class balance
print("Species counts:")
print(df["species"].value_counts())
```

```python
# Filter rows using a condition inside square brackets
# df["species"] == "setosa" creates a True/False series for every row
# df[...] keeps only the rows where the condition is True
setosa = df[df["species"] == "setosa"]
print(f"Setosa rows: {len(setosa)}")

# Select multiple columns by passing a list of column names
# .mean() calculates the average value in each column
print("\nAverage petal dimensions for setosa:")
print(setosa[["petal_length", "petal_width"]].mean().round(2))
```

```python
# Group by species
df.groupby("species")[["sepal_length", "petal_length"]].mean().round(2)
```

## Part 4 — Matplotlib & Seaborn: Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# --- Plot 1: scatter plot coloured by species ---
colors = {"setosa": "#c8401a", "versicolor": "#1a6bc8", "virginica": "#2a8c5a"}
for species, group in df.groupby("species"):
    axes[0].scatter(group["sepal_length"], group["petal_length"],
                    label=species, color=colors[species], alpha=0.7)

axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Petal Length (cm)")
axes[0].set_title("Iris: Sepal vs Petal Length")
axes[0].legend()

# --- Plot 2: petal length distribution per species ---
for species, group in df.groupby("species"):
    axes[1].hist(group["petal_length"], bins=10, alpha=0.6,
                 label=species, color=colors[species])

axes[1].set_xlabel("Petal Length (cm)")
axes[1].set_ylabel("Count")
axes[1].set_title("Petal Length Distribution")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap — which features move together?
plt.figure(figsize=(6, 4))
corr = df.drop(columns="species").corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()

print("\nStrongest correlation:",
      corr.unstack().sort_values(ascending=False).drop_duplicates().iloc[1])
```
