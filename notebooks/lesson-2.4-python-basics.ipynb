## Part 1 — Python Fundamentals

The six constructs you will use in *every* AI script.

```python
# 1. Variables and data types
model_name = "GPT-4"          # string
accuracy   = 0.923            # float
num_params = 175_000_000_000  # int  (underscores for readability)
is_trained = True             # bool

print(f"Model : {model_name}")
print(f"Accuracy : {accuracy:.1%}")
print(f"Parameters: {num_params:,}")
```

```python
# 2. Lists — ordered, mutable sequences
scores = [0.82, 0.91, 0.78, 0.95, 0.88]

print("All scores :", scores)
print("First score:", scores[0])
print("Last score :", scores[-1])
print("Top 3      :", sorted(scores, reverse=True)[:3])
```

```python
# 3. Dictionaries — key-value stores (like a JSON object)
model_info = {
    "name"       : "ResNet-50",
    "task"       : "image classification",
    "top1_acc"   : 0.761,
    "params_M"   : 25.6
}

for key, value in model_info.items():
    print(f"  {key:<12} -> {value}")
```

```python
# 4. Functions — reusable blocks of logic
def accuracy_grade(acc):
    """Return a letter grade for a model accuracy score."""
    if acc >= 0.95:
        return "A+"
    elif acc >= 0.90:
        return "A"
    elif acc >= 0.80:
        return "B"
    else:
        return "C"

for score in scores:
    print(f"  {score:.0%}  =>  {accuracy_grade(score)}")
```

```python
# 5. List comprehensions — compact loops
# Convert fraction accuracies to percentages
pct_scores = [round(s * 100, 1) for s in scores]
print("Percentage scores:", pct_scores)

# Filter only passing scores (>= 85%)
passing = [s for s in pct_scores if s >= 85]
print("Passing (>=85%)  :", passing)
```

```python
# 6. Classes — object-oriented building blocks
class MLModel:
    def __init__(self, name, task):
        self.name = name
        self.task = task
        self.trained = False

    def train(self):
        self.trained = True
        print(f"{self.name} has been trained for '{self.task}'.")

    def predict(self, sample):
        if not self.trained:
            return "Model not yet trained!"
        return f"Prediction on '{sample}' by {self.name}"

my_model = MLModel("TinyBERT", "sentiment analysis")
my_model.train()
print(my_model.predict("I love this AI course!"))
```

## Part 2 — NumPy: Fast Arrays

```python
import numpy as np

# Create arrays
pixels = np.array([120, 200, 85, 255, 30, 178])  # like a row of pixel values
print("Array         :", pixels)
print("Shape         :", pixels.shape)
print("Mean          :", pixels.mean())
print("Std deviation :", pixels.std().round(2))
```

```python
# 2-D arrays — like an image or a matrix of weights
weights = np.array([
    [0.2, -0.5,  0.8],
    [1.1,  0.3, -0.9],
    [0.0,  0.7,  0.4]
])

print("Weights matrix:")
print(weights)
print("\nShape:", weights.shape)   # (rows, columns)
```

```python
# Vectorised math — operates on every element at once (no Python loop needed!)
inputs = np.array([0.5, 1.2, -0.3])

# Matrix multiplication — this is what a neuron does
output = weights @ inputs          # @ is the matrix-multiply operator
print("Neuron outputs:", output.round(3))

# Normalise pixels to 0-1 range (common pre-processing step)
normalised = pixels / 255.0
print("Normalised pixels:", normalised.round(3))
```

```python
# Slicing — extract sub-arrays
matrix = np.arange(1, 26).reshape(5, 5)   # 5x5 grid
print("Full 5x5 matrix:")
print(matrix)

print("\nTop-left 3x3 patch (like a CNN filter sees):")
print(matrix[:3, :3])
```

## Part 3 — pandas: Data Wrangling

```python
import pandas as pd

# Load the classic Iris dataset straight from a URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print(f"Shape: {df.shape}  (rows x columns)")
df.head()
```

```python
# Explore the dataset
print("\n--- Column types ---")
print(df.dtypes)

print("\n--- Summary statistics ---")
df.describe().round(2)
```

```python
# Count species — understanding class balance
print("Species counts:")
print(df["species"].value_counts())
```

```python
# Filter and select
setosa = df[df["species"] == "setosa"]
print(f"Setosa rows: {len(setosa)}")
print("\nAverage petal dimensions for setosa:")
print(setosa[["petal_length", "petal_width"]].mean().round(2))
```

```python
# Group by species
df.groupby("species")[["sepal_length", "petal_length"]].mean().round(2)
```

## Part 4 — Matplotlib & Seaborn: Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# --- Plot 1: scatter plot coloured by species ---
colors = {"setosa": "#c8401a", "versicolor": "#1a6bc8", "virginica": "#2a8c5a"}
for species, group in df.groupby("species"):
    axes[0].scatter(group["sepal_length"], group["petal_length"],
                    label=species, color=colors[species], alpha=0.7)

axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Petal Length (cm)")
axes[0].set_title("Iris: Sepal vs Petal Length")
axes[0].legend()

# --- Plot 2: petal length distribution per species ---
for species, group in df.groupby("species"):
    axes[1].hist(group["petal_length"], bins=10, alpha=0.6,
                 label=species, color=colors[species])

axes[1].set_xlabel("Petal Length (cm)")
axes[1].set_ylabel("Count")
axes[1].set_title("Petal Length Distribution")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap — which features move together?
plt.figure(figsize=(6, 4))
corr = df.drop(columns="species").corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()

print("\nStrongest correlation:",
      corr.unstack().sort_values(ascending=False).drop_duplicates().iloc[1])
```
